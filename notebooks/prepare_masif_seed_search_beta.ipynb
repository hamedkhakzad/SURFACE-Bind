## Preparing the inputs to run MaSIF_seed_search on beta seeds
* There will be two steps to run after this. One on izar and the main one on jed

```python
import json
import os
import pandas as pd
import shutil
```

```python
## read the metrics file to obtain no_binding_sites and list of centroids per protein
table = pd.read_csv('/work/lpdi/users/khakzad/Surfacome/database/result_metrics.csv')
```

```python
accs = table["acc"]
centroids = table["centroids"]
```

```python
# accs_list = json.loads(accs[0])
# centroids_list = json.loads(centroids[0])
for idx, item in enumerate(accs):
    print(item, json.loads(centroids[idx]))
```

```python
import fileinput
import sys

def replaceAll(file,searchExp,replaceExp):
    for line in fileinput.input(file, inplace=1):
        if searchExp in line:
            line = line.replace(searchExp,replaceExp)
        sys.stdout.write(line)
```

```python
dest_path = "/work/lpdi/users/khakzad/masif_seed/masif_seed_search/data/masif_targets/targets/"
template_path = os.path.join(dest_path, "template_beta")
```

```python
for idx, item in enumerate(accs):
    print(item)
    print(json.loads(centroids[idx]))
    site_list = json.loads(centroids[idx])
    try:
        dst = os.path.join(dest_path, item+"_A_beta")
        shutil.copytree(template_path, dst)
        replaceAll(os.path.join(dst,"params_peptides.py"), "#ADD_RESIDUE_LIST", 
              "params['target_residue_list'] = "+str(site_list))
    except(FileNotFoundError):
        continue
```
