## The idea
* The idea is to modify the paramter files with the list of specific residues we want to target.

```python
import fileinput
import sys

def replaceAll(file,searchExp,replaceExp):
    for line in fileinput.input(file, inplace=1):
        if searchExp in line:
            line = line.replace(searchExp,replaceExp)
        sys.stdout.write(line)
```

```python
dest_path = "/work/lpdi/users/khakzad/masif_seed/masif_seed_search/data/masif_targets/targets/"
no_site_path = "/work/lpdi/users/khakzad/masif_seed/masif_seed_search/data/masif_targets/targets/NO_SITE/"
```

```python
for idx, item in enumerate(accs):
    site_list = json.loads(centroids[idx])
    print(item, site_list)
    try:    
        dst1 = os.path.join(dest_path, item+"_A")
        dst2 = os.path.join(dest_path, item+"_A_beta")
        if site_list == []:
            shutil.move(dst1, no_site_path)
            shutil.move(dst2, no_site_path)
        else:
            replaceAll(os.path.join(dst1,"params_peptides.py"), "#ADD_RESIDUE_LIST", 
                "params['target_residue_list'] = "+str(site_list))
 
            replaceAll(os.path.join(dst2,"params_peptides.py"), "#ADD_RESIDUE_LIST", 
                "params['target_residue_list'] = "+str(site_list))
    except(FileNotFoundError):
        print("ERROR: ", item)
        continue
```
